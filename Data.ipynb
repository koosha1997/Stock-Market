# Web crawler ... 

```python
#Fetching price from wallex exchange real time! 
#Import useful libraries ... 
import requests 
from requests.exceptions import ConnectionError
from bs4 import BeautifulSoup
#https://wallex.ir/buy-and-sell/btc

def web_content_div(web_content, class_path):
    web_content_div = web_content.find_all('div',{'class':class_path})
    try:
        spans = web_content_div[0].find_all('span')
        texts = [span.get_text() for span in spans]
    except IndexError:
        texts = []
    return texts

def real_time_price(stock_code):
    url = 'https://wallex.ir/buy-and-sell/' + stock_code
    try: 
        r = requests.get(url)
        web_content = BeautifulSoup(r.text, 'lxml')
        texts = web_content_div(web_content, 'MuiTypography-root MuiTypography-BodyLargeRegular mui-1os5l28')
        if texts != []:
            price = texts[0]
        else:
            price, change = [], []
    except ConnectionError:
        price, change = [], []
    return price

#List of name's stock market that you wan't get data
Stock = 'btc'
print(real_time_price(Stock).replace(',',''))

```

```python
# featching data USDT from abantether ... 
def web_content_div(web_content, class_path):
    web_content_div = web_content.find_all('div',{'class':class_path})
    try:
        buy = int(web_content_div[1].find_all('p')[0].get_text().replace(',', ''))
        sell = int(web_content_div[2].find_all('p')[0].get_text().replace(',', ''))
    except IndexError:
        buy = 0
        sell = 0
    return buy, sell 

def real_time_price(stock_code):
    url = 'https://abantether.com/coin/'+ stock_code
    try: 
        r = requests.get(url)
        web_content = BeautifulSoup(r.text, 'lxml')
        buy, sell = web_content_div(web_content, 'Price_container__DkERw')
        if ((buy != 0)&(sell != 0)):
            print(f'buy : {buy}')
            print(f'sell : {sell}')
    except:
        print('something is wrong!')
        
    return buy, sell
    
 
Stock = 'USDT'
buy,sell = real_time_price(Stock)
```
